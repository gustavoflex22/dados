```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

file_path = list(uploaded.keys())[0]

#carregando o arquivo com a codificação correta
data = pd.read_csv(file_path, encoding='latin1', sep=';')

#convertendo colunas numéricas para o formato correto
for col in ['Compras mensais', 'Número de itens comprados']:
    data[col] = data[col].str.replace(',', '.').astype(float)

#estatísticas descritivas
stats = data.describe()
print("Medidas estatísticas dos dados:")
print(stats)


#histograma de compras mensais
plt.figure(figsize=(8, 5))
sns.histplot(data['Compras mensais'], bins=10, kde=True)
plt.title('Distribuição de compras mensais')
plt.xlabel('Compras Mensais')
plt.ylabel('Frequência')
plt.show()

#histograma de número de itens comprados
plt.figure(figsize=(8, 5))
sns.histplot(data['Número de itens comprados'], bins=10, kde=True)
plt.title('Distribuição do número de itens comprados')
plt.xlabel('Número de itens comprados')
plt.ylabel('Frequência')
plt.show()

#boxplot para detectar outliers
plt.figure(figsize=(8, 5))
sns.boxplot(data=data[['Compras mensais', 'Número de itens comprados']])
plt.title('Boxplot das variáveis numéricas')
plt.show()

#scatter plot para analisar tendências
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=data['Compras mensais'],
    y=data['Número de itens comprados'],
    hue=data['Satisfação com o App (1 a 5)']
)
plt.title('Relação entre compras mensais e número de itens comprados')
plt.xlabel('Compras mensais')
plt.ylabel('Número de itens comprados')
plt.legend(title='Satisfação (1 a 5)')
plt.show()
```
